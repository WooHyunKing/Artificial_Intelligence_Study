```python
import numpy as np
#Weight init

dims=[4096]*7
hs=[]
x=np.random.randn(16,dims[0])
for Din,Dout in zip(dims[0:-1],dims[1:]):
  W=0.01*np.random.randn(Din,Dout)
  print(W.shape)
  x=np.tanh(x.dot(W))
  print(x.shape)
  hs.append(x)
```

```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
for i in range(0,len(hs)):
  H, bins=np.histogram(hs[i])
  ax.plot(bins,np.insert(H,0,H[0]))
ax.legend("012345")
```

```python
import numpy as np
#Xavier init with ReLU

dims=[4096]*7
hs=[]
x=np.random.randn(16,dims[0])
for Din,Dout in zip(dims[0:-1],dims[1:]):
  W=np.random.randn(Din,Dout)*np.sqrt(2/Din)
  print(W.shape)
  x=np.maximum(0,x.dot(W))
  print(x.shape)
  hs.append(x)
```

```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
for i in range(0,len(hs)):
  H, bins=np.histogram(hs[i])
  ax.plot(bins,np.insert(H,0,H[0]))
  ax.legend("012345")
```
